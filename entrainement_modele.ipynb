```python
# -*- coding: utf-8 -*-
"""
# CIFAR-10 Classification avec CNN et PyTorch
Notebook complet pour l'apprentissage et le déploiement d'un modèle CNN sur CIFAR-10

**Objectifs:**
- Comprendre l'architecture CNN pour la classification d'images
- Entraîner un modèle performant sur CIFAR-10
- Sauvegarder le modèle pour utilisation locale
"""

# ============================================================================
# 1. INSTALLATION ET IMPORTS
# ============================================================================

# !pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader

import torchvision
import torchvision.transforms as transforms

import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import time

# Vérification du GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Utilisation du device: {device}")

# ============================================================================
# 2. PRÉPARATION DES DONNÉES
# ============================================================================

"""
CIFAR-10 contient:
- 60000 images couleur 32x32 pixels
- 10 classes: avion, auto, oiseau, chat, cerf, chien, grenouille, cheval, bateau, camion
- 50000 images d'entraînement + 10000 images de test
"""

# Définition des transformations pour augmentation de données
transform_train = transforms.Compose([
    # Remplissage de 4 pixels sur chaque bord
    transforms.RandomCrop(32, padding=4),
    # Retournement horizontal aléatoire (50% de chance)
    transforms.RandomHorizontalFlip(),
    # Conversion en tenseur PyTorch
    transforms.ToTensor(),
    # Normalisation avec moyenne et écart-type de CIFAR-10
    # Ces valeurs sont calculées sur l'ensemble du dataset
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

# Pour les données de test, pas d'augmentation, juste normalisation
transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

# Téléchargement et chargement des datasets
print("Téléchargement du dataset CIFAR-10...")
trainset = torchvision.datasets.CIFAR10(
    root='./data',
    train=True,
    download=True,
    transform=transform_train
)

testset = torchvision.datasets.CIFAR10(
    root='./data',
    train=False,
    download=True,
    transform=transform_test
)

# Création des DataLoaders pour itérer sur les données par batch
batch_size = 128

trainloader = DataLoader(
    trainset,
    batch_size=batch_size,
    shuffle=True,  # Mélange les données à chaque epoch
    num_workers=2  # Nombre de processus pour charger les données
)

testloader = DataLoader(
    testset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=2
)

# Les 10 classes de CIFAR-10
classes = ('avion', 'auto', 'oiseau', 'chat', 'cerf',
           'chien', 'grenouille', 'cheval', 'bateau', 'camion')

print(f"Nombre d'images d'entraînement: {len(trainset)}")
print(f"Nombre d'images de test: {len(testset)}")

# ============================================================================
# 3. VISUALISATION DES DONNÉES
# ============================================================================

def imshow(img):
    """Fonction pour afficher une image"""
    # Dénormalisation de l'image
    img = img / 2 + 0.5
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.axis('off')

# Affichage de quelques images d'exemple
dataiter = iter(trainloader)
images, labels = next(dataiter)

# Afficher 8 images
fig, axes = plt.subplots(2, 4, figsize=(12, 6))
for i, ax in enumerate(axes.flat):
    img = images[i] / 2 + 0.5
    npimg = img.numpy()
    ax.imshow(np.transpose(npimg, (1, 2, 0)))
    ax.set_title(classes[labels[i]])
    ax.axis('off')
plt.tight_layout()
plt.show()

# ============================================================================
# 4. DÉFINITION DE L'ARCHITECTURE CNN
# ============================================================================

class CIFAR10Net(nn.Module):
    """
    Architecture CNN inspirée de VGG mais simplifiée

    Structure:
    - 3 blocs de convolution (chacun avec 2 couches conv + pooling)
    - 3 couches fully connected
    - Dropout pour régularisation
    """

    def __init__(self):
        super(CIFAR10Net, self).__init__()

        # BLOC 1: 3 -> 64 canaux
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(64, 64, kernel_size=3, padding=1)
        self.pool1 = nn.MaxPool2d(2, 2)  # 32x32 -> 16x16
        self.bn1 = nn.BatchNorm2d(64)

        # BLOC 2: 64 -> 128 canaux
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.conv4 = nn.Conv2d(128, 128, kernel_size=3, padding=1)
        self.pool2 = nn.MaxPool2d(2, 2)  # 16x16 -> 8x8
        self.bn2 = nn.BatchNorm2d(128)

        # BLOC 3: 128 -> 256 canaux
        self.conv5 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.conv6 = nn.Conv2d(256, 256, kernel_size=3, padding=1)
        self.pool3 = nn.MaxPool2d(2, 2)  # 8x8 -> 4x4
        self.bn3 = nn.BatchNorm2d(256)

        # Couches fully connected
        # Après 3 poolings: 32/8 = 4, donc 4x4x256 = 4096
        self.fc1 = nn.Linear(256 * 4 * 4, 512)
        self.dropout1 = nn.Dropout(0.5)
        self.fc2 = nn.Linear(512, 256)
        self.dropout2 = nn.Dropout(0.5)
        self.fc3 = nn.Linear(256, 10)  # 10 classes

    def forward(self, x):
        """Passage avant (forward pass)"""
        # Bloc 1
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = self.pool1(x)
        x = self.bn1(x)

        # Bloc 2
        x = F.relu(self.conv3(x))
        x = F.relu(self.conv4(x))
        x = self.pool2(x)
        x = self.bn2(x)

        # Bloc 3
        x = F.relu(self.conv5(x))
        x = F.relu(self.conv6(x))
        x = self.pool3(x)
        x = self.bn3(x)

        # Aplatissement pour les couches fully connected
        x = x.view(-1, 256 * 4 * 4)

        # Couches fully connected avec dropout
        x = F.relu(self.fc1(x))
        x = self.dropout1(x)
        x = F.relu(self.fc2(x))
        x = self.dropout2(x)
        x = self.fc3(x)

        return x

# Création du modèle et déplacement vers GPU si disponible
model = CIFAR10Net().to(device)

# Affichage de l'architecture
print(model)

# Calcul du nombre de paramètres
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nNombre total de paramètres: {total_params:,}")
print(f"Paramètres entraînables: {trainable_params:,}")

# ============================================================================
# 5. DÉFINITION DE LA FONCTION DE PERTE ET DE L'OPTIMISEUR
# ============================================================================

"""
CrossEntropyLoss: combine LogSoftmax et NLLLoss
Idéal pour la classification multi-classes

Adam: optimiseur adaptatif avec momentum
- learning_rate: vitesse d'apprentissage initiale
- weight_decay: régularisation L2 pour éviter l'overfitting
"""

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)

# Scheduler pour réduire le learning rate
# Réduit le LR par 0.1 tous les 30 epochs
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=30, gamma=0.1)

# ============================================================================
# 6. FONCTIONS D'ENTRAÎNEMENT ET D'ÉVALUATION
# ============================================================================

def train_epoch(model, trainloader, criterion, optimizer, device):
    """Entraîne le modèle pendant une epoch"""
    model.train()  # Mode entraînement (active dropout, etc.)
    running_loss = 0.0
    correct = 0
    total = 0

    # Barre de progression
    pbar = tqdm(trainloader, desc='Training')

    for inputs, labels in pbar:
        # Déplacement des données vers GPU
        inputs, labels = inputs.to(device), labels.to(device)

        # Remise à zéro des gradients
        optimizer.zero_grad()

        # Passage avant
        outputs = model(inputs)
        loss = criterion(outputs, labels)

        # Passage arrière et optimisation
        loss.backward()
        optimizer.step()

        # Statistiques
        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

        # Mise à jour de la barre de progression
        pbar.set_postfix({
            'loss': running_loss/len(pbar),
            'acc': 100.*correct/total
        })

    epoch_loss = running_loss / len(trainloader)
    epoch_acc = 100. * correct / total

    return epoch_loss, epoch_acc

def evaluate(model, testloader, criterion, device):
    """Évalue le modèle sur l'ensemble de test"""
    model.eval()  # Mode évaluation (désactive dropout, etc.)
    running_loss = 0.0
    correct = 0
    total = 0

    # Pas de calcul de gradients pour l'évaluation
    with torch.no_grad():
        for inputs, labels in tqdm(testloader, desc='Evaluating'):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    test_loss = running_loss / len(testloader)
    test_acc = 100. * correct / total

    return test_loss, test_acc

# ============================================================================
# 7. EARLY STOPPING
# ============================================================================

class EarlyStopping:
    """
    Arrête l'entraînement si la performance ne s'améliore plus

    Args:
        patience: nombre d'epochs à attendre sans amélioration
        min_delta: amélioration minimale considérée comme significative
    """
    def __init__(self, patience=7, min_delta=0.0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_score = None
        self.early_stop = False

    def __call__(self, val_acc):
        score = val_acc

        if self.best_score is None:
            self.best_score = score
        elif score < self.best_score + self.min_delta:
            self.counter += 1
            print(f"  ⚠️  EarlyStopping counter: {self.counter}/{self.patience}")
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.counter = 0

# ============================================================================
# 8. BOUCLE D'ENTRAÎNEMENT PRINCIPALE
# ============================================================================

# Hyperparamètres
num_epochs = 50  # Maximum d'epochs (early stopping arrêtera avant si nécessaire)
patience = 7     # Nombre d'epochs sans amélioration avant d'arrêter

# Listes pour stocker l'historique
train_losses = []
train_accs = []
test_losses = []
test_accs = []

# Initialisation
best_acc = 0.0
early_stopping = EarlyStopping(patience=patience, min_delta=0.1)

print("\n" + "="*50)
print("DÉBUT DE L'ENTRAÎNEMENT")
print(f"Early Stopping activé (patience={patience})")
print("="*50 + "\n")

start_time = time.time()

for epoch in range(num_epochs):
    print(f"\nEpoch {epoch+1}/{num_epochs}")
    print("-" * 50)

    # Entraînement
    train_loss, train_acc = train_epoch(model, trainloader, criterion, optimizer, device)
    train_losses.append(train_loss)
    train_accs.append(train_acc)

    # Évaluation
    test_loss, test_acc = evaluate(model, testloader, criterion, device)
    test_losses.append(test_loss)
    test_accs.append(test_acc)

    # Mise à jour du learning rate
    scheduler.step()

    # Affichage des résultats
    print(f"\nRésultats epoch {epoch+1}:")
    print(f"  Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
    print(f"  Test Loss:  {test_loss:.4f} | Test Acc:  {test_acc:.2f}%")

    # Sauvegarde du meilleur modèle
    if test_acc > best_acc:
        best_acc = test_acc
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'test_acc': test_acc,
            'test_loss': test_loss,
            'classes': classes,
        }, 'best_model.pth')
        print(f"  ✓ Nouveau meilleur modèle sauvegardé! (Acc: {best_acc:.2f}%)")

    # Vérification early stopping
    early_stopping(test_acc)
    if early_stopping.early_stop:
        print(f"\n🛑 Early Stopping déclenché après {epoch+1} epochs")
        print(f"Pas d'amélioration depuis {patience} epochs")
        break

total_time = time.time() - start_time
print("\n" + "="*50)
print(f"ENTRAÎNEMENT TERMINÉ en {total_time/60:.2f} minutes")
print(f"Nombre d'epochs effectuées: {epoch+1}/{num_epochs}")
print(f"Meilleure précision sur le test: {best_acc:.2f}%")
print("="*50 + "\n")

# ============================================================================
# 9. VISUALISATION DES RÉSULTATS
# ============================================================================

# Graphiques de l'évolution de la perte et de la précision
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Graphique de la perte
ax1.plot(train_losses, label='Train Loss', linewidth=2)
ax1.plot(test_losses, label='Test Loss', linewidth=2)
ax1.set_xlabel('Epoch', fontsize=12)
ax1.set_ylabel('Loss', fontsize=12)
ax1.set_title('Évolution de la perte', fontsize=14, fontweight='bold')
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

# Graphique de la précision
ax2.plot(train_accs, label='Train Accuracy', linewidth=2)
ax2.plot(test_accs, label='Test Accuracy', linewidth=2)
ax2.set_xlabel('Epoch', fontsize=12)
ax2.set_ylabel('Accuracy (%)', fontsize=12)
ax2.set_title('Évolution de la précision', fontsize=14, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# 10. MATRICE DE CONFUSION ET MÉTRIQUES DÉTAILLÉES
# ============================================================================

from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Prédiction sur tout l'ensemble de test
model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for inputs, labels in testloader:
        inputs = inputs.to(device)
        outputs = model(inputs)
        _, preds = outputs.max(1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.numpy())

# Matrice de confusion
cm = confusion_matrix(all_labels, all_preds)

plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=classes, yticklabels=classes,
            cbar_kws={'label': 'Nombre de prédictions'})
plt.title('Matrice de confusion', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Vraie classe', fontsize=12)
plt.xlabel('Classe prédite', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Rapport de classification détaillé
print("\nRAPPORT DE CLASSIFICATION DÉTAILLÉ:")
print("="*70)
print(classification_report(all_labels, all_preds, target_names=classes))

# ============================================================================
# 11. EXEMPLES DE PRÉDICTIONS
# ============================================================================

def predict_image(model, image, label, device):
    """Prédit la classe d'une image"""
    model.eval()
    with torch.no_grad():
        image = image.unsqueeze(0).to(device)
        output = model(image)
        probabilities = F.softmax(output, dim=1)
        confidence, predicted = probabilities.max(1)
    return predicted.item(), confidence.item()

# Affichage de quelques prédictions
dataiter = iter(testloader)
images, labels = next(dataiter)

fig, axes = plt.subplots(3, 4, figsize=(15, 12))
for i, ax in enumerate(axes.flat):
    if i >= len(images):
        break

    img = images[i]
    true_label = labels[i].item()
    pred_label, confidence = predict_image(model, img, true_label, device)

    # Affichage de l'image
    img_display = img / 2 + 0.5
    npimg = img_display.numpy()
    ax.imshow(np.transpose(npimg, (1, 2, 0)))

    # Couleur: vert si correct, rouge si incorrect
    color = 'green' if pred_label == true_label else 'red'
    ax.set_title(f'Vrai: {classes[true_label]}\nPréd: {classes[pred_label]}\nConf: {confidence:.2%}',
                 color=color, fontweight='bold', fontsize=10)
    ax.axis('off')

plt.tight_layout()
plt.show()

# ============================================================================
# 12. SAUVEGARDE FINALE ET CODE POUR UTILISATION LOCALE
# ============================================================================

"""
On garde uniquement le meilleur modèle sauvegardé durant l'entraînement
"""

print("\n" + "="*70)
print("MODÈLE SAUVEGARDÉ:")
print("="*70)
print("✓ best_model.pth - Meilleur modèle durant l'entraînement")
print(f"  - Epoch: {epoch+1}")
print(f"  - Précision: {best_acc:.2f}%")
print("="*70)


# Téléchargement du fichier (pour Google Colab)
try:
    from google.colab import files
    print("\n📥 Téléchargement du modèle...")
    files.download('best_model.pth')
    print("✓ Téléchargement terminé!")
except:
    print("\n💾 Le fichier est sauvegardé dans le répertoire courant")

print("\n" + "="*70)
print("NOTEBOOK TERMINÉ AVEC SUCCÈS!")
print("="*70)
```
